Abaixo está um modulo que usarei para popular a minha tabela no campo de data, o que me ajudou foi [essa](https://pypi.python.org/pypi/RandomWords/0.1.5) informação

```python
from datetime import date
import random

def random_date():
    start_date = date.today().replace(day=1, month=1).toordinal()
    end_date = date.today().toordinal()
    random_day = date.fromordinal(random.randint(start_date, end_date))

    return random_day
```

#### Populando a tabela no banco

```python
for i in range(1000):
    ID = i
    NOME = rw.random_word()
    NASCIMENTO = random_date()
    SALARIO = np.random.randint(1000,size=1)
    
    cursor.execute('insert into cadastro(ID,NOME,NASCIMENTO,SALARIO)' + \
                    'values(?,?,?,?)',
                    (ID,NOME,NASCIMENTO,int(SALARIO)))
```
